> &emsp; O objetivo desta tarefa é mostrar, por meio de um experimento relativamente simples, efeitos caóticos inesperados em modelos matemáticos aparentemente previsíveis. O modelo em questão é o mapa logístico: uma dinâmica discretizada de crescimento populacional.

&nbsp;

> &emsp; Considere $X \in [0, 1]$ como a população teórica do experimento de evolução populacional. Nesse sentido, $X$<sub>0</sub> se refere à população inicial de um experimento, $X$<sub>n</sub> à população do experimento no enésimo passo e $r$ à taxa de crescimento a cada passo. Note que o crescimento será calculado e reportado em função do tempo discretizado de 0 até o número $n$ de passos desejado.

&nbsp;

> &emsp; A equação do mapa logístico consiste em $X$<sub>n+1</sub> $= r * n * (1 - X$<sub>n</sub>$)$. As funções $Logistic$ e $genEvol$ implmentam essa lógica para um número limite de passos (parâmetro "lim").

&nbsp;

> &emsp; Nessa lógica, iremos explorar o comportamento evolutivo do experimento partindo de diferentes valores de $r$ e $X$<sub>0</sub>


$==============================================================================$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Logistic(Xn, r): return r * Xn * (1 - Xn)
```

```python
def genEvol(X0, r, lim=50):
  population = []
  population.append(X0)

  Xn = X0
  n = 1

  while n < lim:
    Xnext = Logistic(Xn, r)
    population.append(Xnext)
    Xn = Xnext
    n += 1

  return population
```

$==============================================================================$

> &emsp; Inicialmente, setamos $X$<sub>0</sub> $= r = 0.5$ e $n = 50$ passos. Consequentemente, por meio do gráfico Tempo vs População, observamos um decrescimento vertiginoso nos primeiros passos do experminento, ultimamente convergindo para 0.

$==============================================================================$


```python
X0 = 0.5; r = 0.5;

time = np.linspace(1, 50, 50)
plt.plot(time, genEvol(X0, r), "r");
plt.xlabel('Discrete Time', fontsize=15);plt.ylabel('Population', fontsize=15);
plt.ylim(-0.01, 1);
```

$==============================================================================$

> &emsp; Nos experimentos a seguir, iremos manter os valores da população inicial e os 50 passos, contudo, para cada valor de $r \in \{0.5, 2.5, 3.1, 3.5, 3.7\}$, o experimento será realizado com valores de $X$<sub>0</sub> $\in \{0.25, 0.75, 0.5, 0.501, 0.5001\}$. Desse modo, podemos visualizar como alterações grandes e pequenas podem influenciar no teor caótico do padrão de crescimento populacional.

$==============================================================================$


```python
r = 0.5

plt.rcParams['figure.figsize'] = [13, 5]

plt.subplot(1, 2, 1);

plt.plot(time, genEvol(0.25, r), "r", label = "X0 = 0.25");

plt.plot(time, genEvol(0.75, r), "g", label = "X0 = 0.75");

plt.ylim(-0.01, 1);
plt.title(f'r = {r}', fontsize=15)
plt.legend(loc = "center", fontsize=12);
plt.xlabel('Discrete Time', fontsize=15);plt.ylabel('Population', fontsize=15);

# ============================================================================ #

plt.subplot(1, 2, 2);

plt.plot(time, genEvol(0.5, r), "r", label = "X0 = 0.5");

plt.plot(time, genEvol(0.501, r), "g", label = "X0 = 0.501");

plt.plot(time, genEvol(0.5001, r), "b", label = "X0 = 0.5001");

plt.ylim(-0.01, 1);
plt.title(f'r = {r}', fontsize=15)
plt.legend(loc = "center", fontsize=12);
plt.xlabel('Discrete Time', fontsize=15);plt.ylabel('Population', fontsize=15);

# ============================================================================ #

plt.show()
```

$==============================================================================$

> &emsp; Para a segunda parte da tarefa, vamos aplicar o mapa logístico em uma escala maior, com o objetivo de explorar o comportamento para um grande intervalo de valores de $r$. Nesse sentido, para cada um dos 10000 valores de $r$ uniformemente distribuídos entre 1 e 4, vamos rodar $genEvol$ 1000 vezes e reportar os últimos 100 valores de população em um gráfico. Portanto, serão mostrados 10.000.000 de valores no gráfico $rvalues$ vs. $population$. Note que fixamos $X$<sub>0</sub> $= 0.5$. O comportamento descrito acima é capturado pela função $longEvol$ a seguir: 

$==============================================================================$


```python
def longEvol():
  r_range = np.linspace(1, 4, 10000)
  x = []; y = [];

  for r in r_range:
    x.append([r for _ in range(100)])
    y.append(genEvol(0.5, r, 1000)[900:])

  return x, y
```

```python
x, y = longEvol()
```

```python
plt.xlabel('r values', fontsize=15)
plt.ylabel('population', fontsize=15)
plt.plot(x, y);
plt.show()
```

$==============================================================================$

> &emsp; Plotando os valores, obtemos o **diagrama de bifurcação**. Para valores de $r$ entre 1 e 3, observamos um crescimento suave na população, indicando comportamento previsível de crescimento a medida que $r$ aumenta. Em $r = 3$, observamos a primeira bifurcação, que, a partir de $r = 3.4$ aproximadamente, torna-se em duas bifurcações. Então, as bifurcações acontecem mais frequentemente e de forma caótica e imprevisível, de modo que a visualização dos valores fica comprometida pela poluição visual provocada.

$==============================================================================$


```python
plt.xlabel('r values', fontsize=15)
plt.ylabel('population', fontsize=15)
plt.plot(x, y);
plt.xlim(3.7, 3.9)
plt.show()
```

$==============================================================================$

> &emsp; Limitando o intervalo de valores de $r$ de 3.7 a 3.9, observamos outro comportamento interessante: dois breves momentos de bifurcações não caóticas aproximadamente ao redor de $r = 3.730$ e de $r = 3.830$. Esses instantes, contudo, são rodeados de oscilações completamente caóticas e dsitribuídas.
